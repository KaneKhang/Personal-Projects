```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Cleaning.csv')
```

```python
df.columns
```

```python
df.head()
```

```python
df.columns = df.iloc[0]
df = df.iloc[1:].reset_index(drop = 1)
```

```python
df.head()
```

```python
pd.set_option('Display.max_rows',500)
```

```python
df1 = df.iloc[:,:8]
df_melt_1 = df1.melt(id_vars = ['Code','Description'], var_name = 'Date', value_name = 'count')
df_melt_1
```

```python
df2 = df.iloc[:,0:2].join(df.iloc[:,9:])
df_melt_2 = df2.melt(id_vars = ['Code','Description'], var_name = 'Date', value_name = 'count')
df_melt_2
```

```python
df_merge = df_melt_1.merge(df_melt_2, on = ['Code','Description','Date'])
df_merge.head()
```

```python
df_merge.columns = ['Code','Description','Date','Units','Values']
df_merge = df_merge.iloc[:,[2,0,1,3,4]]
df_merge
```

```python
df_merge.info()
```

```python
from datetime import datetime as dt
df_merge['Date'] = [dt.strptime(x,'%d.%m.%Y') for x in df_merge['Date']]
```

```python
df_merge['Date'] = [dt.strftime(x,'%d/%m/%Y') for x in df_merge['Date']]
```

```python
df_merge['Date']
```

```python
import statistics as st
df_merge['Units'] = df_merge['Units'].fillna(st.mean(df_merge['Units'].dropna(how = 'any').astype(int)))
```

```python
df_merge['Units']
```

```python
mean = st.mean(df_merge['Values'].str[1:].str.replace(',','').dropna(how= 'any').astype(float))
```

```python
df_merge['Values'] = df_merge['Values'].str[1:].str.replace(',','').fillna(mean).astype(float)
```

```python
df_merge['Values'] = [round(x) for x in df_merge['Values']]
```

```python
df_merge['Units'] = df_merge['Units'].astype(float)
df_merge['Units'] = [round(x) for x in df_merge['Units']]
```

```python
df_merge['Date'] = pd.to_datetime(df_merge['Date'])
```
