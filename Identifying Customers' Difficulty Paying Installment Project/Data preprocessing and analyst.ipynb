# DATA PREPROCESSING

```python
#Check blank values
for i in df.columns:
    print(f'------------------\n{i}\n',df.loc[df[i] == ''].value_counts())
```

```python
#Check duplicates
df.duplicated().sum()
```

```python
#Check NaN values for category
for i in category:
    print(f'------------------\n{i}\n',df[i].isna().value_counts())
```

```python
#Check NaN values for numeric
for i in numeric:
    print(f'------------------\n{i}\n',df[i].isna().value_counts())
```

```python
#Fix NaN values for category
for i in category:
    if df[i].isna().any():
        df[i].fillna('Unknown',inplace = True)
```

```python
df[category].isnull().any()
```

```python
#Download for further analyst in Power BI
df.to_csv('data.csv')
```

```python
#Remove outliers of DAYS_EMPLOYED column
df = df.loc[df['DAYS_EMPLOYED'] != 365243]
df.shape
```

```python
df['TARGET'].value_counts()
```

```python
#Fix NaN values for numeric (for correlation)
import statistics
for i in numeric:
    if df[i].isna().any():
        df[i].fillna(df[i].dropna(how='any').mean(),inplace = True)
```

```python
df[numeric].isnull().any()
```

```python
# Encode categorical features using the get_dummies function
df = pd.get_dummies(df, columns = category)
```

```python
df.drop(numeric,axis = 1).head()
```

```python
#Show all results in output
pd.set_option('display.max_rows', 500)
```

```python
#Check correlation toward target column
print(df.drop(['SK_ID_CURR'],axis = 1).corrwith(df['TARGET']))
```

```python
X = df.drop(['TARGET','SK_ID_CURR'],axis = 1)
y = df['TARGET']
```

```python
from sklearn.model_selection import train_test_split
# Split dataset into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 128) # %70 train data, %30 test data

# Print lengths of training and testing datasets for each feature and target variable
print("X_train: ",len(X_train))
print("X_test: ",len(X_test))
print("y_train: ",len(y_train))
print("y_test: ",len(y_test))
```

```python
#Feature importance
from IPython.core.display import display_markdown
from pandas.core.dtypes.common import conversion
from sklearn.ensemble import RandomForestClassifier

tree = RandomForestClassifier(random_state=128)

tree.fit(X_train,y_train)

feat_imp = pd.DataFrame(tree.feature_importances_, index=X_train.columns, columns = ['importance']).sort_values('importance',ascending = False)
feat_imp
```

```python
plt.figure(figsize=(11,9))
feat_imp = pd.Series(tree.feature_importances_, index=X_train.columns)
chart=feat_imp.nlargest(10).sort_values().plot(kind='barh')
plt.show()
```
