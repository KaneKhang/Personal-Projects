```python
#Transform values for heatmap 
df['Target'] = df['Target'].replace('Graduate',2).replace('Dropout',0).replace('Enrolled',1)
```

```python
df = df.astype('float')
df = df.astype('int32')
```

```python
plt.figure(figsize=(13,9))
sns.heatmap(df.corr(),vmax=0.8,linewidth=0.1,cmap='coolwarm')
plt.show()
```

```python
#prepare for Important Feature
x = df.drop('Target',axis=1)
y = df.Target
```

```python
from sklearn.ensemble import ExtraTreesClassifier

extra_tree = ExtraTreesClassifier()

extra_tree.fit(x,y)

plt.figure(figsize=(11,9))
feat_imp = pd.Series(extra_tree.feature_importances_, index=x.columns)

chart=feat_imp.nlargest(10).sort_values().plot(kind='barh', color=['grey','grey','grey','grey','grey','blue','red','yellow','orange','green'])
plt.show()
```
